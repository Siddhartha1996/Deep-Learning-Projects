```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(-1,1))
data = sc.fit_transform(data)
```

```python
data[:10]
```

```python
look_back = 5
dataX, dataY = [],[]
for i in range(len(data)-(look_back)):
    dataX.append(data[i:i+look_back])
    dataY.append(data[i+look_back])
```

```python
dataX = np.asanyarray(dataX)
dataY = np.asanyarray(dataY)
```

```python
dataX.shape
```

```python
dataY.shape
```

```python

xtrain = dataX[:50,:,:]
xtest = dataX[50:,:,:]
ytrain = dataY[:50]
ytest = dataY[50:]
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape, ytest.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM,RepeatVector
```

```python
from keras.layers import Flatten,Conv1D, MaxPooling1D
```

```python
model = Sequential()
model.add(Conv1D(filters =256,kernel_size=2, activation='relu', input_shape=(5,1)))
model.add(Conv1D(filters=128, kernel_size=2,activation='relu', input_shape=(5,1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(RepeatVector(30))
model.add(LSTM(128,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
history = model.fit(xtrain,ytrain,epochs=100,verbose=1)
```

```python
pred = model.predict(xtest)
pred = sc.inverse_transform(pred)
```
