```python
df = df.drop(['sl_no','salary'],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
le=LabelEncoder()
```

```python
df['gender'] = le.fit_transform(df['gender'])
df['ssc_b'] = le.fit_transform(df['ssc_b'])
df['hsc_b'] = le.fit_transform(df['hsc_b'])
df['hsc_s'] = le.fit_transform(df['hsc_s'])
df['degree_t'] = le.fit_transform(df['degree_t'])
df['workex'] = le.fit_transform(df['workex'])
df['specialisation'] = le.fit_transform(df['specialisation'])
df['status'] = le.fit_transform(df['status'])
```

```python
df
```

```python
y=df.status
X =df.drop(['status'],axis=1)
```

```python
y
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=20)
```

```python


sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
model = Sequential()
model.add(Dense(128, init='uniform', activation='softmax', input_dim=X_train.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(64, init='uniform', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, init='uniform', activation='sigmoid'))
model.add(Dropout(0.2))
model.add(Dense(4,init='uniform',activation='softmax'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.add(Dropout(0.2))
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train, validation_data=(X_test,y_test), nb_epoch=100,batch_size=400)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=(y_pred>0.5)
```

```python
cm=confusion_matrix(y_test,y_pred)
fig,ax=plt.subplots(figsize=(5,5))
ax.matshow(cm,cmap=plt.cm.Blues,alpha=0.7)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(x=j,y=i,s=cm[i,j],va='center',ha='center')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
