```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler(feature_range=(0,1))
df_URBN=sc.fit_transform(df_URBN)
```

```python
df_URBN
```

```python
#Splitting training and test data:


train_size = int(len(df_URBN) * 0.8)
test_size = len(df_URBN) - train_size
train, test = df_URBN[0:train_size, :], df_URBN[train_size:len(df_URBN), :]
```

```python
def create_dataset( dataset, look_back ):
    dataX, dataY =[],[]
    for i in range( len(dataset) - look_back - 1 ):
        a = dataset[ i:( i+look_back ), 0 ]
        dataX.append(a)
        dataY.append(dataset[i+look_back,0])
    return np.array(dataX), np.array(dataY)
```

```python
# Reshape into X=t,Y=t+1
look_back=90
xtrain,ytrain=create_dataset(train, look_back)
xtest,ytest=create_dataset(test, look_back)
xtrain=np.reshape(xtrain, (xtrain.shape[0],xtrain.shape[1],1))
xtest=np.reshape(xtest, (xtest.shape[0],xtest.shape[1],1))
```

```python
print(train)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.layers import LSTM
```

```python
from keras.layers import Activation
```

```python
#Building the model:

model=Sequential()

model.add(LSTM(input_shape=(xtrain.shape[1],1), units=50, return_sequences=True)) # adding LSTM layer
model.add(Dropout(0.2)) #adding Dropout

model.add(LSTM(100, return_sequences = False))                            #Adding LSTM layer
model.add(Dropout(0.2))                                                   #Adding Dropout
model.add(Dense(units=1))                                                 #Adding Dense layer with activation = "linear"
model.add(Activation('linear'))
```

```python
model.compile(optimizer='adam',loss="mean_squared_error")
model.summary()
```

```python
history=model.fit(xtrain,ytrain,epochs=200,batch_size=50,validation_data=(xtest,ytest),verbose=1)
```

```python
prediction=model.predict(xtest)
```

```python
prediction=sc.inverse_transform(prediction)
ytest=sc.inverse_transform(ytest)
```

```python
ax=[x for x in range(180)]
plt.figure(figsize=(8,4))
plt.plot(ax, ytest[0][:180], marker='.', label="actual")
plt.plot(ax, prediction[:,0][:180], 'r', label="prediction")
plt.tight_layout()
plt.subplots_adjust(left=0.07)
plt.ylabel('Price', size=15)
plt.xlabel('Time step', size=15)
plt.legend(fontsize=15)
plt.show();
```
