```python
X = df.drop(['target'],axis=1)
```

```python
y = df['target']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier

#Model:

```

```python
from keras.layers import Dense, Dropout

model=Sequential()
model.add(Dense(1024,activation='relu',init='uniform',input_dim=X_train.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu',init='uniform'))
model.add(Dropout(0.2))
model.add(Dense(128,activation='softmax',init='uniform'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
model.add(Dropout(0.2))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train, validation_data=(X_test,y_test), nb_epoch=200,batch_size=50)
```

```python
y_pred=model.predict(X_test)
```
