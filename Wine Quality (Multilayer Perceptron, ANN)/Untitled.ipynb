```python
y = df[12]
```

```python
y
```

```python
y = pd.get_dummies(y)
```

```python
y
```

```python
X = df.iloc[:,0:12]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=0)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
y_train
```

```python
y_test
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
model = Sequential()
model.add(Dense(128, init='uniform', activation='softmax', input_dim=X_train.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(64, init='uniform', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32,init='uniform',activation='softmax'))
model.add(Dropout(0.2))
model.add(Dense(7, activation='sigmoid'))
model.add(Dropout(0.2))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    batch_size=500,
                    epochs=50,
                    verbose=2,
                    validation_data=(X_test, y_test))
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
y_test=y_test.to_numpy()
```

```python
y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)
```

```python
print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```

```python
df_result = pd.DataFrame.from_dict(history.history)
```

```python
df_result['accuracy'].plot()
```

```python

df_result['loss'].plot()
```

```python
cm=confusion_matrix(y_test_class,y_pred_class)
fig,ax=plt.subplots(figsize=(5,5))
ax.matshow(cm,cmap=plt.cm.Blues,alpha=0.6)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(x=j,y=i,s=cm[i,j],va='center',ha='center')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
